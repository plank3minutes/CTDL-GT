```python
import pandas as pd
import random

# Ngân hàng câu nhận xét
comments = {
    'T': [
        "Rất xuất sắc, hãy tiếp tục phát huy!",
        "Bạn đã hoàn thành vượt kỳ vọng.",
        "Điểm số cao, chúc mừng bạn!",
        "Thể hiện rất tốt, tiếp tục nhé!",
        "Kết quả tuyệt vời, hãy tự hào!"
    ],
    'H': [
        "Bạn đã hoàn thành tốt, cần cố gắng thêm!",
        "Điểm số ổn định, tiếp tục nỗ lực nhé!",
        "Hoàn thành nhiệm vụ, hãy cố hơn nữa.",
        "Chúc mừng bạn hoàn thành mục tiêu!",
        "Cần cải thiện nhưng bạn đã hoàn thành tốt!"
    ],
    'C': [
        "Cần cải thiện nhiều để đạt mục tiêu.",
        "Bạn chưa đạt kỳ vọng, hãy cố gắng!",
        "Nên ôn tập thêm để cải thiện điểm.",
        "Cần sự nỗ lực hơn để đạt kết quả tốt.",
        "Không đạt yêu cầu, cần chú trọng hơn!"
    ]
}



# Hàm tính toán điểm trung bình cho các cột
def calculate_common_ability_score(row):
    scores = [row['Tự chủ và tự học'], row['Giao tiếp và hợp tác'], row['Giải quyết vấn đề và sáng tạo']]
    return sum(scores) / len(scores)

def calculate_specialized_ability_score(row):
    scores = [
        row['Ngôn ngữ'], row['Tính toán'], row['Khoa học'], row['Công nghệ'], row['Tin học'],
        row['Thẩm mĩ'], row['Thể chất']
    ]
    return sum(scores) / len(scores)

def calculate_character_score(row):
    scores = [row['Yêu nước'], row['Nhân ái'], row['Chăm chỉ'], row['Trung thực'], row['Trách nhiệm']]
    return sum(scores) / len(scores)

# Tạo nhận xét từ điểm trung bình
def assign_comment(score):
    if score >= 9:
        level = "T"
    elif score >= 5:
        level = "H"
    else:
        level = "C"
    return random.choice(comments[level])  # Chọn ngẫu nhiên nhận xét

# Lặp qua các sheet trong nhóm 'NL-PC'
for sheet_name in sheet_names['NL-PC']:
    # Đọc dữ liệu từ sheet
    df = pd.read_excel(file_path, sheet_name=sheet_name)

    # Đảm bảo ngày sinh ở dạng dd/MM/yyyy
    if 'Ngày sinh' in df.columns:
        df['Ngày sinh'] = pd.to_datetime(df['Ngày sinh'], errors='coerce').dt.strftime('%d/%m/%Y')

    # Tính và gán nhận xét cho các cột
    df['Nhận xét năng lực chung'] = df.apply(calculate_common_ability_score, axis=1).apply(assign_comment)
    df['Nhận xét năng lực đặc thù'] = df.apply(calculate_specialized_ability_score, axis=1).apply(assign_comment)
    df['Nhận xét phẩm chất'] = df.apply(calculate_character_score, axis=1).apply(assign_comment)

    # Append vào file Excel "Final", mỗi sheet có tên ứng với tên sheet hiện tại
    with pd.ExcelWriter(final_file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
        df.to_excel(writer, index=False, sheet_name=sheet_name)

    print(f"Sheet {sheet_name} đã được append vào file Final.xlsx")
```
